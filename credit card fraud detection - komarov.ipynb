# Anomaly detection

```python
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import GridSearchCV
from sklearn.covariance import EllipticEnvelope
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
```

#### GridSearchCV should use everywhere scoring=my_auc_score defined below.

Y_hat is the **score**, showing how much an object looks like an outlier. So **predict** methods should **return scores, not classes.**

To obtain such predict method, you will need to redefine (by inheritance) original scikit-learn methods.

Don't confuse object outlier score with object normality score.

```python
from sklearn.metrics import make_scorer

def my_auc(Y,Y_hat):
    fpr,tpr,_ = roc_curve(Y, Y_hat, pos_label=1)
    return auc(fpr, tpr)

my_auc_score = make_scorer(my_auc, greater_is_better=True)
```

#### Show quality on test set.

### EmpiricalCovariance method

1. fit the unsupervised model to TEST SET (you can do that since it is unsupervised)
1. show ROC curve for TEST SET with title 'AUC=<value>', where <value> is estimated AUC value.

```python
from sklearn.covariance import EmpiricalCovariance
```

```python
ec2 = EmpiricalCovariance()
ec2.fit(X_test)
dist = ec2.mahalanobis(X_test)
#нормируем расстояние
predict = (dist - min(dist)) / (max(dist) - min(dist))
auc_value = my_auc(Y_test, predict)
fpr,tpr,_ = roc_curve(Y_test, predict, pos_label=1)
```

```python
print("AUC_VALUE = {}".format(auc_value))
plt.plot(fpr, tpr)
plt.title("ROC")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.show()
```

### IsolationForest method

1. find optimal parameters with GridSearchCV with default number of folds on TRAIN SET (set n_jobs=1 (GridSearchCV may not work otherwise)).
1. display best parameters

1. fit the unsupervised model to TEST SET (you can do that since it is unsupervised)
1. predict TEST SET using best model hyperparameters from TRAIN SET cross-validation
1. show ROC curve for TEST SET with title 'AUC=<value>', where <value> is estimated AUC value.

Plot the graph how the quality depends on **n_estimators**.

```python
class IsolationForest2(IsolationForest):    
    def predict(self, X):
        func =  -self.decision_function(X)
        return (func - min(func)) / (max(func) - min(func))
```

```python
isolation_forest = IsolationForest2()
param_grid = {'n_estimators':[5, 10, 20, 30, 50, 100]}
clf = GridSearchCV(isolation_forest, param_grid, scoring = my_auc_score)
clf.fit(X_train, Y_train)
```

```python
print("best params: {}".format(clf.best_params_))
```

```python
print("best score: {}".format(clf.best_score_))
```

```python
isolation_forest2 = clf.best_estimator_
```

```python
isolation_forest2.fit(X_test.values)
prob = isolation_forest2.predict(X_test)
fpr,tpr,_ = roc_curve(Y_test, prob, pos_label=1)
```

```python
print'AUC = {}'.format(my_auc(Y_test, prob))
plt.plot(fpr, tpr)
plt.title("ROC")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.show()
```

```python
n_estimators = [5, 10, 20, 30, 50, 100]
y = clf.cv_results_['mean_test_score']
plt.plot(n_estimators, y)
plt.xlabel("n_estimators")
plt.ylabel("AUC")
plt.show()
```
